```python
#dep time arrival time var bunları da yapıcaz
train_data.head()
```

```python
train_data['Journey_day']=pd.to_datetime(train_data['Date_of_Journey'],format="%d/%m/%Y").dt.day
```

```python
train_data['Journey_month']=pd.to_datetime(train_data['Date_of_Journey'],format="%d/%m/%Y").dt.month
```

```python
train_data.head()
```

```python
#2 tane değerli veriyi aldık artık object tipi olan data of joruneye gerek kalmadı
train_data.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
train_data.head()
```

```python
train_data['Dep_Hour']=pd.to_datetime(train_data['Dep_Time']).dt.hour
```

```python
train_data['Dep_Min']=pd.to_datetime(train_data['Dep_Time']).dt.minute
```

```python
train_data.head()
```

```python
train_data.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
train_data.head()
```

```python
#arrival time için de aynısını yapak
#extract hours
train_data["Arrival_Hour"]=pd.to_datetime(train_data['Arrival_Time']).dt.hour
```

```python
train_data["Arrival_Minute"]=pd.to_datetime(train_data['Arrival_Time']).dt.minute
```

```python
train_data.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
train_data.head()
```

```python
"2h 50m".split()
```

```python
len("2h 50m".split())
```

```python
# Time taken by plane to reach destination is called Duration
# It is the differnce betwwen Departure Time and Arrival time


# Assigning and converting Duration column into list
duration = list(train_data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration
```

```python
train_data['Duration_hours']=duration_hours
train_data['Duration_mins']=duration_mins
```

```python
train_data.head()
```

```python
#categorical data
#nominal data ->data are not in any order -> one hot encoder
#ordinal data->data are in order  -> LabelEncoding  #sıralı degılse
```

```python
#airlineyi one hot encoding yapıcaz
Airline=train_data[['Airline']]
Airline=pd.get_dummies(Airline,drop_first=True)
Airline.head()
```

```python
#sourceyi onehotencoding yap
Source=train_data[['Source']]
Source=pd.get_dummies(Source,drop_first=True)
Source.head()
```

```python
train_data['Destination'].value_counts()
```

```python
Destination=train_data[['Destination']]
Destination=pd.get_dummies(Destination,drop_first=True)
Destination.head()
```
