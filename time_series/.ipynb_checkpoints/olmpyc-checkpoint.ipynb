```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
athlets=pd.read_csv('athlete_events.csv')
region=pd.read_csv('noc_regions.csv')
```

```python
region.head()
```

```python
atlet_df=athlets.merge(region,how='left',on='NOC')
atlet_df.head()
```

```python
top_10_country=atlet_df.Team.value_counts().sort_values(ascending=False).head(10)
```

```python
top_10_country
```

```python
plt.figure(figsize=(12,6))
plt.title('Overall Country')
sns.barplot(x=top_10_country.index,y=top_10_country,palette='Set2')
```

```python
plt.figure(figsize=(12,6))
plt.title('yaşlar')
plt.xlabel('Age')
plt.ylabel('Katılımcı sayısı')
plt.hist(atlet_df.Age,bins=np.arange(10,80,2),color='orange',edgecolor='white')
```

```python
gender_counts=atlet_df.Sex.value_counts()
gender_counts
```

```python
#gender çizdir
plt.figure(figsize=(12,6))
plt.title('Cinsiyet Dağılımı')
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=150,shadow=True)
```

```python
atlet_df['Medal'].value_counts()
```

```python
women=atlet_df[(atlet_df['Sex']=='F')&(atlet_df['Season']=='Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=women,palette='Spectral')
plt.title('Kadın katılımı')
```

```python
part=women.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Kadınların katılım grafiği')
```
